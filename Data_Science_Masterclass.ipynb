```python
import pandas as pd
```

```python
df=pd.read_csv("retail.csv")
```

```python
df
```

```python
df.head(5)
```

```python
df.tail(5)
```

```python
df[['Profit','Sales']].describe()
```

```python
#average sales value per order

sales_per_order = df.groupby('Order ID')['Sales'].mean()
```

```python
sales_per_order.mean()
```

```python
#average sales value per state
df.groupby('State')['Sales'].mean().sort_values(ascending=False)
```

```python
df.groupby('State')['Sales'].sum().sort_values(ascending=False)
```

```python
df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
```

```python
df.groupby('Category')['Sales'].sum().sort_values(ascending=False).plot()
```
